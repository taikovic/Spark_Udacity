```python
import pyspark
from pyspark import SparkConf
from pyspark.sql import SparkSession
```

```python
spark = SparkSession \
    .builder \
    .appName("Our first Python Spark SQL example") \
    .getOrCreate()
```

```python
spark.sparkContext.getConf().getAll()
```

```python
spark
```

```python
path = "./posts.json"
user_log = spark.read.json(path)
```

```python
user_log.printSchema()
```

```python
user_log.describe()
```

```python
user_log.show(1)
```

```python
user_log.select("title").take(5)
```

```python
user_log.take(5)
```

```python
out_path = "./posts_formatted.csv"
```

```python
user_log.write.save(out_path,format="csv",header = True)
#we should remove non string columns
```
